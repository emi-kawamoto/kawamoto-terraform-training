# オブジェクト指向

## オブジェクト指向とは

プログラムの記述、設計の際に用いられる考え方の一つです。

変数と関数をひとまとめにした**クラス**を作る点が特徴です。

しかし、オブジェクト指向には正解と呼べる形がなく、また難しい概念ですのでここでは完璧な理解をしようとしなくても大丈夫です。

実際にプログラムを書くときや設計するときの考え方、メリットを認識することをゴールとします。

## クラスとオブジェクト

オブジェクト指向のコンテキストとして**クラス**と**オブジェクト**があります。

### クラス

クラス(Class)とは、「概念を定義したもの」です。

以下の画像の例は、「人」という概念とその概念が持つデータを定義しています。

この、「その概念が持つデータ」を**コンストラクタ引数**といいます。



````python
# イメージ画像を表示
from IPython.display import Image
Image("imgs/class_human.png")
````



具体的なデータなしに、その概念だけを定義したものがクラスです。

## オブジェクト

しかし、概念だけでは「食べる」「歩く」などの処理を実行できません。

処理を実行するためには概念を実体化させることが必要です。

この実体化させることを**インスタンス化**といい、インスタンス化されたものを**オブジェクト**(インスタンス)といいます。


````python
# イメージ画像を表示
from IPython.display import Image
Image("imgs/tarohanako.png")
````


## なぜオブジェクト指向を使うのか

### 保守がしやすい(継続的に管理しやすい)

オブジェクト指向では、概念の振る舞いを**メソッド**として定義することができます。

例えば、人であれば「食べる」「歩く」などの振る舞いを共通で持っています。


````python
# イメージ画像を表示
from IPython.display import Image
Image("imgs/method.png")
````



この場合、`taro`も`hanako`も「食べる」「歩く」という処理を実行することができます。

もし途中で振る舞いを追加したい場合には、クラスにメソッドを追加するだけで、人クラスをもつ全てのオブジェクトは追加された振る舞いをすることができます。

### 分割しやすい

複数人での開発などでは、開発部分を分割しなくてはいけない場合があります。

オブジェクト指向を使うと、オブジェクトの振る舞いの詳細を気にしなくても実装が可能なので開発やテスト時の分割、結合が容易になります。

## オブジェクト指向のメリット/デメリット

### メリット
* 慣れれば分かりやすい
  * 独立しているので全体を読み解く必要がない
* 開発を効率的にできる
  * 個々の機能毎に担当者を決めることで、並行して開発がしやすい（大規模開発で特に向いている）
* 再利用ができる
  * 処理単位でプログラムが作られるので、他のプロジェクトでも再利用がしやすい

### デメリット
* 考え方や用語（クラス・インスタンス・カプセル化・継承・ポリモーフィズム など）の理解が必要で、勉強をせずに誰でも使えるものではない
  * 逆に「オブジェクト指向がわかる」というのは、エンジニアとしてのステータスにもなります


# Pythonでオブジェクト指向プログラミング

## クラスの定義

猫の概念である、`Cat`クラスを定義してみます。

````python
class Cat():
    def __init__(self, name, kind, weight):
        self.name = name
        self.kind = kind
        self.weight = weight
````

`__init__()` は、インスタンス化された時に自動的に実行されるメソッドです(**コンストラクタ**)。

これがあることで、インスタンス化される時に渡されるデータをメソッド内で使用できるようになります。

### コンストラクタ引数とプロパティ

コンストラクタに渡される引数のことを**コンストラクタ引数**といい、

```python
        self.name = name
        self.kind = kind
        self.weight = weight
```

のようにしてメソッド内で扱えるように定義されたデータ(変数)を**プロパティ**といいます。

## インスタンス化

`Cat`クラスをインスタンス化してみます。

````python
tama = Cat("tama", "マンクス", 4.5)
maru = Cat("maru", "スコティッシュ・フォールド", 4)
````

## メソッドの作成

作成した`Cat`クラスに、メソッド`eat`を追加します。

````python
class Cat():
    def __init__(self, name, kind, weight):
        self.name = name
        self.kind = kind
        self.weight = weight

    def eat(self, food):
        self.weight += 0.2
        print("{0}は{1}を食べて体重が {2}kg になりました".format(self.name, food, self.weight))
````

````python
tama = Cat("tama", "マンクス", 4.5)
tama.eat("チュール")
````

### `self`

`self`は、自分自身を指します。

上記の例では、`eat`メソッドを呼び出した`tama`オブジェクトのプロパティに対する操作となります。

## 継承

継承とは、**「すでに定義しているクラスをもとに、新しく要素を加えたクラスを定義すること」**です。

親となるクラスを**スーパークラス**、子となるクラスを**サブクラス**と呼んだりします。

````python
# イメージ画像を表示
from IPython.display import Image
Image("imgs/inheritance.png")
````

````python
class Cat():
    def __init__(self, name, kind, weight):
        self.name = name
        self.kind = kind
        self.weight = weight

    def eat(self, food):
        self.weight += 0.2
        print("{0}は{1}を食べて体重が {2}kg になりました".format(self.name, food, self.weight))

class Scottish(Cat):
    def __init__(self, name, kind, weight, earFeatures):
        super().__init__(name, kind, weight)
        self.earFeatures = earFeatures

    def earBend(self):
        print("耳を曲げました！")
````

上記の例では、`Cat`クラスを継承した`Scottish`クラスを定義しています。

```python
class Scottish(Cat):
```

のように、括弧の中に継承元(親)のクラス名を書くことで継承します。

```python
super().__init__(name, kind, weight)
```

`super().処理名` でメソッド内で継承元のメソッドを呼び出すことができます。

継承をすることで、新たな要素(耳の形・耳を曲げる動作)を追加しながら、継承元のプロパティやメソッドが使用できます。


````python
maru = Scottish("maru", "スコティッシュ・フォールド", 4, "垂れ耳")
````

````python
maru.earBend()
````

````python
# 継承元のメソッドも使える
maru.eat("まぐろ")
````

### オーバーライド

サブクラスでは、継承元と同じ名前のメソッドを作成することができます。

メソッドを呼び出した時、新しく作成されたサブクラスでのメソッドが優先されます。

同じ名前を設定して上書きすることをオーバーライドといいます。


````python
class Scottish(Cat):
    def __init__(self, name, kind, weight, earFeatures):
        super().__init__(name, kind, weight)
        self.earFeatures = earFeatures

    # 継承元と同じ名前のメソッドを設定
    def eat(self):
        print("チュールばっかり食べています")
````

````python
maru = Scottish("maru", "スコティッシュ・フォールド", 4, "垂れ耳")
# 新たに設定したメソッドが優先される
maru.eat()
````

### ポリモーフィズム(多態性)

メソッドのオーバーライドをすることで、スーパークラスの振る舞いをサブクラス固有のものとして定義し直すことができます。

このように、「同じ命令を送ったにもかかわらず、それぞれが独立した固有の処理を行うという特性」のことを、**ポリモーフィズム**といいます。

### 継承のメリット

* コードの追加・修正といったメンテナンスが容易になる
* より抽象的な共通する機能をまとめておくことができる

また、オーバーライド・ポリモーフィズムのメリットは「別の名前で似たような処理のメソッドが乱立しなくなる（可読性が高まる）」点があります。

## カプセル化

カプセル化とは、オブジェクトの情報を1つにまとめ、外部からデータや処理に対してアクセスできないようにすることです。

クラス外部から、

* クラス自体
* プロパティ
* メソッド

を呼び出せないようにすることで実現できます。

### メリット

* 一度関数を通す状況を作ることで不正なデータ操作を防ぐことができる
  * 関数で条件や設定を設ける！
* 呼び出し側が内部の実装を気にしなくても良い

````python
class Cat():
    def __init__(self, name, kind, weight):
        self.name = name
        self.kind = kind
        self.weight = weight
        self.__bank = 0

    def get_bank(self):
        return self.__bank

    def set_bank(self, bank):
        self.__bank = bank

    def __hello(self):
        print("Hello!!!")

    def print_hello(self):
        self.__hello()

tama = Cat("tama", "マンクス", 4.5)
````

### 変数のカプセル化

```python
 def __init__(self, name, kind, weight):
    self.__bank = 0
```    

`__`(アンダースコア2つ)を先頭につけるとプライベート変数になり、外部からの呼び出し・変更ができなくなります。

````python
# 失敗する（プライベート変数なので外部から参照はできない）
tama.__bank
````

### ゲッターとセッターを使って値を操作する

プライベートな変数を操作したい場合は、ゲッターやセッターの役割を持つメソッドを作成してやる必要があります。

値の操作に必ず関数を必要とするため、関数内で値の妥当性の確認やエラー処理をすることができます。

```python
def get_bank(self): #値を取得するメソッド(ゲッター)
    return self.__bank
def set_bank(self, bank): #値を設定するメソッド(セッター)
    self.__bank = bank
```

````python
tama.set_bank(10000)
tama.get_bank()
````

### メソッドのカプセル化

```python
def __hello(self):
    print("Hello!!!")
def print_hello(self):
    self.__hello()
```

変数と同様に`__`をメソッド名の先頭につけることでプライベートメソッドとして定義することができます。

このクラスの中でしか許されない処理などはプライベートにすることができます。

以下の例だと、__hello を直接呼び出すことはできません。

クラス内のメソッドである print_hello() からは呼び出すことはできます。

````python
# 失敗する（プライベートメソッドなので外部から実行はできない）
tama.__hello()
````

````python
# 成功する
tama.print_hello()
````
